```python
import numpy as np
import faiss
import requests
from io import StringIO
import pandas as pd

from io import BytesIO
```

```python
# URLs of the .npy embedding files on GitHub
urls = [
    'https://raw.githubusercontent.com/bonsonsm/FAISS/master/02_embeddings/bert/embeddings_0.npy',
    'https://raw.githubusercontent.com/bonsonsm/FAISS/master/02_embeddings/bert/embeddings_1.npy',
    # Add more URLs as needed
]
```

```python
def load_npy_from_github(url, ctr):
    str_extention = '.npy'
    new_url = url+str(ctr)+str_extention
    print(new_url)

    # Fetch the file from the raw GitHub link
    response = requests.get(new_url)
    response.raise_for_status()  # Check if the request was successful

    # Load the .npy file into a NumPy array
    np_array = np.load(BytesIO(response.content))
    return np_array
```

```python
def create_final_embedding(base_url, num_files):

  # Load each embedding and store in a list
  embeddings = [load_npy_from_github(base_url, ctr) for ctr in range(num_files)]

  # Merge all the loaded embeddings along the first axis (row-wise)
  merged_embeddings = np.concatenate(embeddings, axis=0)

  print("Merged Embedding Shape:", merged_embeddings.shape)
  return merged_embeddings
```

```python

path_bert_str = 'https://raw.githubusercontent.com/bonsonsm/FAISS/master/02_embeddings/bert/embeddings_'
path_albert_str = 'https://raw.githubusercontent.com/bonsonsm/FAISS/master/02_embeddings/albert/embeddings_'

#Total embeddings are 57 - From 0 to 56
total_embeddings = 57

bert_embedding = create_final_embedding(path_bert_str, total_embeddings)
albert_embedding = create_final_embedding(path_albert_str, total_embeddings)
```

```python
dimensions_bert = bert_embedding.shape[1]  # columns
print(dimensions_bert)
index_bert = faiss.IndexFlatL2(dimensions_bert)
index_bert.is_trained
```

```python
dimensions_albert = albert_embedding.shape[1]  # columns
print(dimensions_albert)
index_albert = faiss.IndexFlatL2(dimensions_albert)
index_albert.is_trained
```

```python
index_bert.add(bert_embedding)
index_albert.add(albert_embedding)
```

```python
print(index_bert.ntotal)
print(index_albert.ntotal)
```

**Search given a query xq and number of nearest neigbors to return k.**

```python
from sentence_transformers import SentenceTransformer
model_bert = SentenceTransformer('bert-base-nli-mean-tokens')
#model_albert = SentenceTransformer('paraphrase-albert-small-v2')

xq_sentence = "Bonson likes to play the guitar"

k = 4
xq_bert = model_bert.encode([xq_sentence])
#xq_albert = model_albert.encode([xq_sentence])
```

```python
import requests

def read_raw_text():
# Read Sentences from Github
  sentences=[]

  # URL of the raw text file from GitHub
  url = 'https://raw.githubusercontent.com/bonsonsm/FAISS/master/01_data/collated_similarity_sentences.txt'

  # Send a GET request to fetch the raw content
  response = requests.get(url)

  # Check if the request was successful
  if response.status_code == 200:
      # Split the content by lines and store it in a list
      sentences = response.text.splitlines()
      # Print the list to verify
      # print(sentences[0])
  else:
      print(f"Failed to retrieve file: {response.status_code}")

  return sentences
```

```python
sentences = read_raw_text()
```
